El **HDR o High Dynamic Range**, és un conjunt de tècniques de processat d'imatges que serveix per a obtenir imatges amb un **rang de lluminositat més ampli** en comparació amb el que es pot aconseguir amb imatges estàndar o tècniques fotogràfiques.

L'ull humà té la **capacitat d'adaptar-se a l'entorn**, fent que, independentment de la lluminositat de la situació en la que es troba, adaptant la mesura de l'iris, l'huma pugui processar els colors de forma constant. Això permet als humans veure un rang ampli de condicions llumíniques.

L'objectiu del HDR és **adaptar el rang de lluminositat de les imatges**, acostant-lo al rang que té la vista humana.



Les càmeres que avui en dia no utilitzen cap tècnica de HDR, solen perdre informació en els detalls de brillantors o ombres.

Hi ha dos grups principals de tècniques de HDR:
- **Render amb ordinador:** es realitza un processat de les imatges amb ordinador per a obtenir un millor rang de lluminositat.


- **Merge de LDR:** s'utilitza un conjunt de imatges amb un Low Dynamic Range, fusionant-les per a obtenir una sola imatge sense pèrdua de informació(aproximació).

També es pot aconseguir fent servir diferents tipus de sensors fotogràfics.


Per a representar imatges HDR, s'ha de tenir en compte que el rang de lluminositat ha de poder fer-se visible en els monitors i televisors d'avui en dia. Els televisors HDR, per exemple, fan servir una tècnica anomenada *Tone Mapping* la qual consisteix en reduïr el contrast d'una imatge HDR per tal que les imatges puguin ser representades en un dispositiu que no pot mostrar un rang tant ampli de lluminositat.


------



Les fotografies, en les cameres digitals, solen ser capturades utilitzant matrius de 24 bits, fent servir 8 bits per canal de color, on cada canal prendrà un valor de 0 a 255. Això implica que la captura de llum que pot realitzar una càmera digital és limitat, numèricament.

La tècnica més comú de HDR és la que s'utilitza en telèfons mòvils, on es realitzen imatges en successió molt ràpida(per a evitar el moviment entre captures), on cada imatge es captura fent servir temps d'exposició diferents. Combinant la informació d'aquestes 3 imatges es pot obtenir una imatge HDR.


Fent servir python i OpenCV podem aplicar aquesta tècnica de forma relativament senzilla:

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread
```

### Obtenir imatges amb diferents temps d'exposició

Necessitem 3 imatges:
- **Una imatge sotaexposada:** imatge més fosca, per a capturar els detalls de la imatge en parts molt brillants.
- **Una imatge amb una exposició adient:** imatge regular, basada en la iluminació de l'escena.
- **Una imatge sobreexposada:** imatge més brillant, per a capturar els detalls de la imatge en parts fosques.

Per a fer el següent exemple hem utilitzat un telèfon mòvil amb el mode manual, fent tres captures amb temps d'exposició diferents(0.25, 2.5 i 15.0), al ser imatges realitzades manualment el resultat segurament presentarà soroll, però ens serveix per a exemplificar processos HDR(seguint els tutorials de openCV).

```python
def readImagesAndTimes():
    # List of exposure times
    times = np.array([0.25, 2.5, 15.0 ], dtype=np.float32)

    # List of image filenames
    filenames = [ "img/exp_025.jpeg", "img/exp_250.jpeg", "img/exp_1500.jpeg"]
    images = []
    for filename in filenames:
        im = cv2.imread(filename)
        images.append(im)

    return images, times
images, exp_times = readImagesAndTimes()
```

```python
low_exp = imread("img/exp_025.jpeg")
mid_exp = imread("img/exp_250.jpeg")
high_exp = imread("img/exp_1500.jpeg")


plt.figure(figsize=(15,15))
plt.subplot(131)
plt.title("Low exposure image")
plt.imshow(low_exp)
plt.subplot(132)
plt.title("Medium exposure image")
plt.imshow(mid_exp)
plt.subplot(133)
plt.title("High exposure image")
plt.imshow(high_exp)
```

### Fusió de les imatges
Ajuntem la seqüència d'imatges amb diferents temps d'exposició formant una sola imatge HDR. Utilitzarem dos mètodes diferents que proporciona OpenCV, Debvec i Robertson.
Aquets algoritmes utilitzen els temps d'exposició de les imatges per a realitzar la fusió.

```python
# Merge exposures to HDR image
merge_debvec = cv2.createMergeDebevec()
hdr_debvec = merge_debvec.process(images, times=exp_times.copy())
merge_robertson = cv2.createMergeRobertson()
hdr_robertson = merge_robertson.process(images, times=exp_times.copy())
```

### *Tonemapping*
Com s'ha explicat anteriorment, per a poder veure les imatges en un monitor cal passar les dades de la imatge HDR a una nova imatge amb valors entre 0 i 1.

```python
# Tonemap HDR image
tonemap1 = cv2.createTonemapDrago(gamma=1.0, saturation=1.0, bias=0.86)
res_debvec = tonemap1.process(hdr_debvec.copy())
tonemap2 = cv2.createTonemapDrago(gamma=1.0, saturation=1.0, bias=0.86)
res_robertson = tonemap2.process(hdr_robertson.copy())
```

### Mertens fusion
Com a alternativa als algoritmes Debvec i Robertson, es pot utilitzar una tècnica de fusió com Mertens, aquesta tècnica no necessita els temps d'exposició ni realitzar tonemapping sobre el resultat d'aquesta.

```python
# Exposure fusion using Mertens
merge_mertens = cv2.createMergeMertens()
res_mertens = merge_mertens.process(images)
```

### Convertir les dades a 8bit i guardar

```python
# Convert datatype to 8-bit and save
res_debvec_8bit = np.clip(res_debvec*255, 0, 255).astype('uint8')
res_robertson_8bit = np.clip(res_robertson*255, 0, 255).astype('uint8')
res_mertens_8bit = np.clip(res_mertens*255, 0, 255).astype('uint8')

cv2.imwrite("ldr_debvec.jpg", res_debvec_8bit)
cv2.imwrite("ldr_robertson.jpg", res_robertson_8bit)
cv2.imwrite("fusion_mertens.jpg", res_mertens_8bit)
```

# Resultats
Per a l'anàlisis d'aquests resultats cal remarcar que les imatges que hem utilitzat les hem realitzat amb el nostre propi telèfon(OnePlus 3), utilitzant el mode manual en una habitació amb molt poca llum, Entenent que la òptica utilitzada per capturar les imatges és molt bàsica(no es tracta d'una càmara reflex).

```python
result_debvec = imread("ldr_debvec.jpg")
result_robertson = imread("ldr_robertson.jpg")
result_mertens = imread("fusion_mertens.jpg")
image_auto =  imread("auto_op.jpeg")
```

A continuació podem veure una imatge capturada amb el mode automàtic del OnePlus 3, amb el mode HDR activat. Ens servirà per a poder comparar els resultats obtinguts fent servir les captures manuals.

```python
plt.figure(figsize=(15,15))
plt.title("Room image using auto-mode")
plt.imshow(image_auto)
```

#### Debvec
Els resultats obtinguts utilitzant Debvec són bsatant positius, hi ha una mica de soroll en la part més fosca de les imatges(la part inferior dreta) però creiem que això és degut a les impresicions de la captura feta amb el telèfon mòvil.

Hem obtingut una imatge bastant realitsta de l'habitació, on es poden observar detalls al vidre de la porta, alhora que es veuen detalls de la part fosca(l'arbre de nadal i la paret).

Cal afegir que hi ha una mica de moviment entre les 3 captures originals, fent que el resultat sigui una mica borrós(no hem aplicat un mètode de *image matching* per a fer-ho més simple)

```python
plt.figure(figsize=(15,15))
plt.title("Result using Debvec")
plt.imshow(result_debvec)
```

#### Robertson
Els resultats obtinguts utilitzant Robertson no són bons. No sabem si és degut als paràmetres utilitzats al fer *tone mapping* o degut a la imprecisió de les imatges originals i els temps d'exposició. La majoria de tutorials utilitzen un toneMapping diferent, però aquest no està disponible en python així que deixem els resultats obtinguts amb Drago.

```python
plt.figure(figsize=(15,15))
plt.title("Result using Robertson")
plt.imshow(result_robertson)
```

#### Mertens
La tècnica de fusió alternativa ha donat uns resultats realment bons. La imatge conté molta informació tant de les parts fosques com les brillants.

Considerem que el resultat obtingut és millor que el que proporciona el propi telèfon movil que hem utilitzat, ja que a la imatge original no es poden apreciar tants detalls a la part de la paret i l'arbre de nadal.

Hi ha una mica més de soroll degut al procés i també, repetim, les imatges que hem realitzat manualment tenen un petit *displacement* el qual no hem corregit, de forma que s'afegeix borrositat(apreciable als vidres de la porta).

```python
plt.figure(figsize=(15,15))
plt.title("Result using Mertens")
plt.imshow(result_mertens)
```
